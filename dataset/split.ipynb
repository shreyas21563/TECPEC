```python
import json
```

```python
train_val_data = json.load(open('Subtask_1_train.json'))
test_data = json.load(open('Subtask_1_test.json'))
train_data = train_val_data[:int(len(train_val_data)*0.9)]
val_data = train_val_data[int(len(train_val_data)*0.9):]
```

```python
json.dump(train_data, open('train.json', 'w'))
json.dump(val_data, open('val.json', 'w'))
json.dump(test_data, open('test.json', 'w'))
```
